```python
import numpy
from numpy import linspace, zeros, sqrt, arctan2, sin, cos
from matplotlib import pyplot as plt
```

```python
def F(x, y, kx=1, ky=1):
    # theta = arctan2(y, x)
    # f = - k * sqrt(x ** 2 + y ** 2)
    # return f * cos(theta), f * sin(theta)
    return - kx * x, - ky * y
```

```python
m = 1 # kg
kx = sqrt(7) # N / m
ky = sqrt(5) # N / m

N = 9000
to = 0
tf = 30
dt = (tf - to) / N
t = linspace(to, tf, N + 1)

x = zeros(t.shape)
y = zeros(t.shape)

vx = zeros(t.shape)
vy = zeros(t.shape)

ax = zeros(t.shape)
ay = zeros(t.shape)

x[0] = 0 # m
y[0] = 0 # m

vx[0] = 1  # m/s
vy[0] = 1  # m/s
```

```python
# i = 0
# print("{0:5d}  {1:10.4f}  {2:10.4f}  {3:10.4f}  {4:10.4f}  {5:10.4f}"\
#     .format(i, time[i], x[i], y[i], vx[i], vy[i]))

for i, _ in enumerate(t[1:] , 1):
    vx[i] = vx[i-1] + ax[i-1] * dt
    vy[i] = vy[i-1] + ay[i-1] * dt

    x[i] = x[i-1] + vx[i] * dt
    y[i] = y[i-1] + vy[i] * dt

    ax[i], ay[i] = F(x[i], y[i], kx, ky)
    ax[i], ay[i] = ax[i] / m, ay[i] / m

    # print("{0:5d}  {1:10.4f}  {2:10.4f}  {3:10.4f}  {4:10.4f}  {5:10.4f}"\
    #     .format(i, time[i], x[i], y[i], vx[i], vy[i]))
```

```python
plt.plot(
    t, 0.5 * m * (vx * vx + vy * vy), 'r-',
    t, 0.5 * kx * x * x + 0.5 * ky * y * y, 'g-.',
    t, 0.5 * m * (vx * vx + vy * vy) + 0.5 * kx * x * x + 0.5 * ky * y * y, 'b--',
)
```
